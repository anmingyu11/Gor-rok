# $plog(p)$

$$
y = plog(p) \\
p \in [0,1]
$$

```python
x = np.linspace(0.001, 0.999, 1000)
```

```python
y = -x * np.log2(x)
```

```python
plt.figure(figsize=(12,8))
plt.plot(x,y)
plt.show()
```

```python
x[np.argmax(y)]
```

# $-\sum_i{p_ilogp_i} \ i \in [0,1] $

```python
p1 = np.linspace(0.001, 0.999, 1000)
p0 = 1-p1
ent = -(p1*np.log2(p1) + p0* np.log2(p0))
```

```python
plt.figure(figsize=(12,8))
plt.plot(p1,ent)
plt.show()
```

# $-\sum_i{p_ilogp_i} \ i \in [0,n] $

$n$  $entropy$ 

```python
ps = []
ents = []
for i in range(1, 50):
    p = 1 / i
    ent = -i * p * np.log2(p)
    ps.append(p)
    ents.append(ent)
```

```python
ps[10:30]
```

```python
ents[10:30]
```

```python
plt.figure(figsize=(12,8))
plt.plot(ps, ents)
plt.show()
```
